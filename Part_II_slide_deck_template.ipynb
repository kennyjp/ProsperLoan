## Aperçu de l'ensemble de données


```python
# importer tous les packages et définir les tracés à intégrer en ligne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# supprimer les avertissements de la sortie finale
import warnings
warnings.simplefilter("ignore")
```

```python
# charger l'ensemble de données dans une base de données pandas
df = pd.read_csv('prosperLoanData-clean.csv')
```

## Visualisation 1

> Il est à priorie nécéssaire de connaitre Quel "Duree" compte le plus grand nombre de prêts dans l'ensemble de données ? afin de voir sur quel période l'analyse est fondé et comment?


```python
# Créer un diagramme circulaire pour montrer la distribution
plt.figure(figsize=[8,8])
c = df.Duree.value_counts()
labels = '36Mois','60Mois','12Mois'
plt.pie(c, autopct='%1.1f%%', startangle=90)
plt.title('Distribution de fréquence absolue de la durée du prêt')
plt.axis('square')
plt.legend(labels);
```

> D'après le graphique ci-dessus, nous pouvons voir que la majorité des personnes qui ont contracté des prêts ont opté pour le plan de 36 mois. Donc une proportion plus élevée de prêts ont été approuvés pour une période de 36 mois, tandis qu'une proportion moindre a été approuvée pour une période de 60 mois avec une très faible proportion de prêts approuvés pour une période de 12 mois.Ceci nous permet de voir déjà les délais viable de prêt pour être accepter.

## Visualisation 2

> Quelle relation ou modèle existe-t-il entre les variables numériques de l'ensemble de données favorisant le prêt ?

```python
# liste de toutes les variables numériques
numeric_var = ['Duree','MontantInitial','Interet','MensualitéPret','DureeStatut',
               'DebtToIncomeRatio','RevenuMensuel','TotalCreditLinespast7years']
# correlation plot
plt.figure(figsize = [8, 6])
sb.heatmap(df[numeric_var].corr(), annot = True, fmt = '.3f', center = 0)
plt.title('Corrélation des variables numériques grace à la carte thermique');
plt.show();
```

>le taux emprunteur a une forte corrélation. Le taux emprunteur augmentera si le TAEG augmente, et inversement. Les investisseurs et le montant initial du prêt ont une corrélation modérée. Donc la populations semble satisfaite de ce modéle d'aide financier et est aussi crédible sur l'échéance

## Visualisation 3

> Examinons la relation entre Interet, LoanStatus et la Duree du Prêt ? afin de permettre une meilleur corrélation et apporter de bonne conclusion d'analyse

```python
# Création d'une boîte à moustaches pour les variables multivariées de BorrowerAPR, LoanStatus et Term avec le terme comme codage couleur
plt.figure(figsize=[20,8])
sb.boxplot(data = df, y = 'Interet', x = 'LoanStatus', hue = 'Duree')
plt.legend(loc = 6, bbox_to_anchor = (1.0, 0.5)) # legende de la figure
plt.xticks(rotation = 15);
```

> La catégorie de statut de prêt en défaut a un aperçu intéressant ici, montrant que les prêts à terme de 12 mois ont un Intéret sur l'emprunteur plus élevé. Cette observation n'était pas perceptible lorsque le graphique bivarié de LoanStatus vs Duree a été précédemment tracé, ce qui donne un excellent aperçu de ce graphique. Aussi Defaulted, Changedoff, PastDue(1-15days), PastDue(16-30days),PastDue(61-90days), PastDue(31-60days), PastDue(91-120days), PastDue(>120 days) montrent une tendance similaire avec une durée de 12 mois qui n'était pas perceptible dans le graphique bivarié de LoanStatus vs Duree.
