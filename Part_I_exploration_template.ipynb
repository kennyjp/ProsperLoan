#### Question 1 : Quelle est la distribution de la variable `LoanStatus` et laquelle d'entre elles a le nombre le plus élevé ?

```python
# Définir la couleur par défaut de la palette de couleurs
base_color = sb.color_palette()[0]
# Utilisation de Countplot pour la distribution
data_order = df.LoanStatus.value_counts().index
sb.countplot(data = df, y = 'LoanStatus', color = base_color, order = data_order)
plt.title('Counts For Each Loan Status');
plt.ylabel('Loan Status');
```

#### Observations
> Il est évident dans la visualisation ci-dessus que le statut de prêt 'Current' qui veut dire Actuel a le nombre le plus élevé dans la distribution.
> Le LoanStatus montre un aperçu plutôt intéressant, ce qui signifie qu'une grande majorité des prêts ont été complétés tandis qu'une petite proportion de prêts ont été soit en défaut de paiement, soit en retard de plusieurs jours sur la période de remboursement. Cependant, étant donné qu'un grand nombre de prêts accordés (plus de 50 %) sont toujours en cours d'exécution sur la base des données disponibles, nous n'étudierons que les facteurs qui affectent le statut du prêt qui sont soit terminés, en défaut ou en souffrance.

#### Question 2 : Quel "Duree" compte le plus grand nombre de prêts dans l'ensemble de données ?

```python
# Obtenons la valeur compte de chaque Durrée
df.Duree.value_counts()
```
